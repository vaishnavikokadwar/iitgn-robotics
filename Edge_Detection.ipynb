```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

Sobel Edge Detection

```python
image = cv2.imread("1637062780706.jpg") #reading the image
cv2_imshow(image) #show original image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #convert to grayscale
plt.imshow(gray, cmap="gray")
plt.show()
edges = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) #applying Sobel edge detection
cv2_imshow(edges) #show filtered image
```

Canny Edge Detection

```python
edges = cv2.Canny(image=gray, threshold1=100, threshold2=200) #apply Canny edge detection filter
# Display Canny Edge Detection Image
cv2_imshow(edges)
```
