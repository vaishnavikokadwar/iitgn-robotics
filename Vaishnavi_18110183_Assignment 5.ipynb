Task 2: Stanford Manipulator

```python
def dyn(n,D,phi):

  c = [[[0]*n]*n]*n

  #calculating Christoffer symbols
  for k in range(n):
    for j in range(n):
      for i in range(n):
        c[i][j][k] = 0.5*(sp.diff(D[k][j],q[i])+sp.diff(D[k][i],q[j])-sp.diff(D[i][j],q[k]))
        
  tau = [0]*n #torque for each link

  d = 0
  ct = 0
  for k in range(n): #corresponds to each link

    for j in range(n):
      d = d + D[k][j] * qddot[j]
      
      for i in range(n):
        ct = ct + c[i][j][k]*qdot[i]*qdot[j]
    tau[k] = d + ct + phi[k]

  for i in range(n):
    print('tau'+str(i+1)+'='+str(tau[i]))
  return(tau)
```

```python
def inverse_kinematics(endeffector_position,lengthsoflinks):
    theta1 = np.arctan(endeffector_position[1]/endeffector_position[0])
    r = np.sqrt(endeffector_position[0]**2 + endeffector_position[1]**2)
    s = endeffector_position[2] - lengthsoflinks[0]
    theta2 = np.arctan(s/r)
    d3 = np.sqrt(r**2 + s**2) - lengthsoflinks[1]
    return theta1, theta2, d3
    # print("First Solution: \n", "Theta1 = ", theta1, "\n Theta2 =", theta2,"\n Extension: ", d3, "\n")
    # print("Second Solution: \n", "Theta1 = ", np.pi + theta1, "\n Theta2 =", np.pi - theta2,"\n Extension:", d3)
```

```python
#user inputs
l1 = 10
l2 = 10
m1 = 5
m2 = 5
m3 = 5
initial_position = [10,12,5] 
final_position = [5,10,5] 

Kp = 10
Ki = 10

[theta1_d, theta2_d, d3_d] = inverse_kinematics(final_position,[l1,l2]) #desired trajectory joint variables
[theta1_i, theta2_i, d3_i] = inverse_kinematics(initial_position,[l1,l2]) #initial joint variables

[theta1_int, theta2_int, d3_int] = inverse_kinematics(initial_position,[l1,l2]) #intermediate joint variables

#intial error values
prev_e1 = theta1_d - theta1_int
prev_e2 = theta2_d - theta2_int
prev_e3 = d3_d - d3_int
tstep = 0.01 #time step

plt.ion()
plt.show()

while (theta1_int != theta1_d) or (theta2_int != theta2_d) or (d3_int != d3_d): #PI angle control for first joint variable
  e1 = theta1_d - theta1_int
  tau1 = Kp*e1 + Ki*(e1-prev_e1)*(tstep) 

  e2 = theta2_d - theta2_int
  tau2 = Kp*e2 + Ki*(e2-prev_e2)*(tstep)

  e3 = d3_d - d3_int
  tau3 = Kp*e3 + Ki*(e3-prev_e3)*(tstep)

  o0 = [0,0,0] #origin and first joint
  o1 = [0,0,l1] #second joint
  o2 = [l2*np.sin(theta1_int),l2*np.sin(theta1_int),l1] #third joint
  o3 = [l2*np.sin(theta1_int)+d3_int*np.sin(theta2_int),l2*np.cos(theta1_int), l1+d3_int*np.cos(theta2_int)] #end effector position

  plt.plot([o0[0],o1[0]],[o0[1],o1[1]],[o0[2],o1[2]],'-o') #plotting the mechanism
```

Scara Manipulator

```python
def scara_invkin(x,y,z,d1,d2):
    # using formulae from the textbook
    r = abs((x**2+y**2-d1**2-d2**2)/(2*d1*d2))
    theta2 = np.arctan(np.sqrt(abs(1-r**2))/r)
    theta1 = np.arctan(y/x) - np.arctan((d2*np.sin(theta2))/(d1+d2*np.cos(theta2)))
    d3 = -z
    # converting from radians to degrees
    theta1 = theta1*180/np.pi
    theta2 = theta2*180/np.pi
    return theta1, theta2, d3
```
