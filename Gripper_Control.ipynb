```python
import numpy as np
import matplotlib.pyplot as plt
```

PD Control

```python
m = 5 #mass of gripper
g = 9.8

d_i = 15 #distance between two parts of gripper initially according to width and length of the pill
d_f = 0 #final distance between two parts of gripper when two parts come together to lock the pill

d_int = d_i #intermediate distance
ddot_int = 0 #initial velocity

prev_err = d_f - d_i #initial value of the dynamic error
tstep = 0.001 #time step
Kp = 0.1
Kd = 0.1

t = 0 #total time required

plt.ion()
plt.show()

while d_int > d_f+0.1 or d_int < d_f-0.1:
  err = d_int - d_f
  force = Kp*err + Kd*(prev_err-err)/tstep

  #from dynamic equation for a 1DOF gripper we get the following
  ddot_int = (ddot_int + (force/m + g)*tstep)
  d_int = (d_int + (force/m - g)*tstep**2 + ddot_int*tstep)

  t = t +1

  #print(d_int)
  plt.clf()

  plt.xlim([-1, 35])
  plt.ylim([-1, 45])

  plt.plot([0,0],[0,43.215],'-o')
  plt.plot([d_int,d_int+18.641],[0,23.215],'-o')
  plt.plot([d_int+18.641,d_int+18.641],[23.215,43.215],'-o')

  plt.pause(0.0001)

plt.ioff()
plt.show()
print('Total time required for the operation:' + str(t*tstep) + 'seconds')
```

PID Control

```python
m = 5 #mass of gripper
g = 9.8

d_i = 15 #distance between two parts of gripper initially according to width and length of the pill
d_f = 0 #final distance between two parts of gripper when two parts come together to lock the pill

d_int = d_i #intermediate distance
ddot_int = 0 #initial velocity

prev_err = d_f - d_i #initial value of the dynamic error
tstep = 0.001 #time step
Kp = 0.1
Kd = 0.1
Ki = 0.1

t = 0

plt.ion()
plt.show()

while d_int > d_f+0.1 or d_int < d_f-0.1:
  err = d_int - d_f
  force = Kp*err + Kd*(prev_err-err)/tstep + Ki*(prev_err - err)*tstep

  ddot_int = (ddot_int + (force/m + g)*tstep)
  d_int = (d_int + (force/m - g)*tstep**2 + ddot_int*tstep)

  t = t + 1

  #print(d_int)
  plt.clf()

  plt.xlim([-1, 35])
  plt.ylim([-1, 45])

  plt.plot([0,0],[0,43.215],'-o')
  plt.plot([d_int,d_int+18.641],[0,23.215],'-o')
  plt.plot([d_int+18.641,d_int+18.641],[23.215,43.215],'-o')

  plt.pause(0.0001)

plt.ioff()
plt.show()
print('Total time required for the operation:' + str(t*tstep) + 'seconds')
```
